# 이미지파일 픽셀로 변환하기

```python
from PIL import Image
import numpy as np
import os.path

# covid 환자가 아닌경우
targetdir = r"COVID_dataset/curated_data/curated_data/1NonCOVID" #해당 폴더 설정 

files = os.listdir(targetdir)

format = [".png"] #지원하는 파일 형태의 확장자들

y_data=np.array([])

count=0
for (path,dirs,files) in os.walk(targetdir):
    for file in files:
        image = Image.open(path+"\\"+file)
        print(count)
        img_resize=image.resize((224, 224))
        pix=np.array(img_resize)
        y_data=np.append(y_data,1)
        if count==0:
            x_data=pix
        else:
            x_data = np.concatenate((x_data, pix), axis=0)
        count++
        
# covid 환자인 경우
targetdir = r"COVID_dataset/curated_data/curated_data/2COVID" #해당 폴더 설정 

files = os.listdir(targetdir)

format = [".png"] #지원하는 파일 형태의 확장자들

for (path,dirs,files) in os.walk(targetdir):
    for file in files:
        image = Image.open(path+"\\"+file)
        print(count)
        img_resize=image.resize((224, 224))
        pix=np.array(img_resize)
        y_data=np.append(y_data,2)
        x_data = np.concatenate((x_data, pix), axis=0)   
        
# 폐렴환자인 경우
targetdir=r"COVID_dataset/curated_data/curated_data/3CAP"

files = os.listdir(targetdir)

format = [".png"] #지원하는 파일 형태의 확장자들

for (path,dirs,files) in os.walk(targetdir):
    for file in files:
        image = Image.open(path+"\\"+file)
        print(count)
        img_resize=image.resize((224, 224))
        pix=np.array(img_resize)
        y_data=np.append(y_data,3)
        x_data = np.concatenate((x_data, pix), axis=0)   
```
